```python
# Activity 2: Visual Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Load data again
df = pd.read_csv('../data/liver_dataset_cleaned.csv')

# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot for class distribution
sns.countplot(x='Target', data=df)
plt.title("Target Class Distribution (0 = Healthy, 1 = Cirrhosis)")
plt.xlabel("Liver Disease Status")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of a few key features
features_to_plot = ['Total_Bilirubin', 'Direct_Bilirubin', 'Alkaline_Phosphotase', 'Albumin']

for feature in features_to_plot:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Boxplots to detect outliers
for feature in features_to_plot:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Target', y=feature, data=df)
    plt.title(f'{feature} vs Target')
    plt.show()
```
